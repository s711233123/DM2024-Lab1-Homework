````python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

#Create separate DataFrames for each category
categories = X['sentiment'].unique()  # Get unique category labels
category_dfs = {}  # Dictionary to store DataFrames for each category

for category in categories:
    # Filter the original DataFrame by category
    category_dfs[category] = X[X['sentiment'] == category].copy()

# Function to create term-document frequency DataFrame for each category
def create_term_document_df(df):
    count_vect = CountVectorizer()  # Initialize the CountVectorizer
    X_counts = count_vect.fit_transform(df['comment'])  # Transform the text data into word counts
    
    # Get the unique words (vocabulary) from the vectorizer
    words = count_vect.get_feature_names_out()
    
    # Create a DataFrame where rows are documents and columns are words
    term_document_df = pd.DataFrame(X_counts.toarray(), columns=words)
    
    return term_document_df

# Create term-document frequency DataFrames for each category
filt_term_document_dfs = {}  # Dictionary to store term-document DataFrames for each category

for category in categories:
    filt_term_document_dfs[category] = create_term_document_df(category_dfs[category])
````

````python
category_number=0 #You can change it from 0 to 3
word_counts = filt_term_document_dfs[categories[category_number]].sum(axis=0).to_numpy()

# Sort the term frequencies in descending order
sorted_indices = np.argsort(word_counts)[::-1]  # Get indices of sorted frequencies
sorted_counts = np.sort(word_counts)[::-1]  # Sort frequencies in descending order

# Calculate the index corresponding to the top 5% most frequent terms
total_terms = len(sorted_counts)
top_5_percent_index = int(0.05 * total_terms)

# Get the indices of the top 5% most frequent terms
top_5_percent_indices = sorted_indices[:top_5_percent_index]

# Filter terms that belong to the top 5% based on their rank
filtered_words = [filt_term_document_dfs[categories[category_number]].iloc[:, i].name for i in top_5_percent_indices]

print(f"Category: {categories[category_number]}")
print(f"Number of terms in top 5%: {top_5_percent_index}")
print(f"Filtered terms: {filtered_words}")
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Calculate word frequencies from TF-IDF matrix
def calculate_word_frequencies(tfidf_matrix, feature_names):
    return dict(zip(feature_names, tfidf_matrix.sum(axis=0).A1))

# Generate word cloud
def create_wordcloud(word_freq, width=800, height=400, bg_color='white', colormap='viridis'):
    wordcloud = WordCloud(width=width, height=height, background_color=bg_color, colormap=colormap).generate_from_frequencies(word_freq)

    # Display word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Plot top N words
def plot_top_tfidf_words(word_freq, top_n=20, figsize=(12, 8)):
    # Sort and select the top N words
    sorted_word_scores = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:top_n]
    words, scores = zip(*sorted_word_scores)

    # Plot bar chart
    plt.figure(figsize=figsize)
    plt.barh(words, scores, color='skyblue')
    plt.xlabel('TF-IDF Score')
    plt.ylabel('Words')
    plt.title(f'Top {top_n} Words by TF-IDF Score')
    plt.gca().invert_yaxis()
    plt.show()

# Calculate word frequencies
word_freq = calculate_word_frequencies(y, vectorizer.get_feature_names_out())

# Generate word cloud with specific colors
create_wordcloud(word_freq, bg_color='black', colormap='plasma')

# Plot top 20 words by TF-IDF score
plot_top_tfidf_words(word_freq)
````

````python
import numpy as np
import plotly.express as px

def plot_top_terms(tfidf_matrix, vectorizer, top_n=50, frequency_threshold=1):
    # Calculate term frequencies
    term_frequencies = tfidf_matrix.sum(axis=0).A1  # Flatten to 1D array

    # Filter terms based on frequency threshold
    indices = np.where(term_frequencies > frequency_threshold)[0]
    terms = vectorizer.get_feature_names_out()[indices]
    filtered_frequencies = term_frequencies[indices]

    # Sort terms and select top N
    sorted_indices = np.argsort(filtered_frequencies)[::-1][:top_n]
    terms_sorted = terms[sorted_indices]
    frequencies_sorted = filtered_frequencies[sorted_indices]

    # Apply logarithmic transformation
    frequencies_log = np.log(frequencies_sorted)

    # Plot with Plotly
    fig = px.bar(x=terms_sorted, y=frequencies_log, 
                 labels={'x': 'Terms', 'y': 'Log Frequency'},
                 title=f'Top {top_n} Terms by Log Frequency')
    fig.show()

# Example usage
plot_top_terms(y, count_vect)
````

## Section 4

In the lab, we quickly applied each step to illustrate how to work with the dataset. However, some aspects of this process were not ideal, efficient, or meaningful. Each dataset should be handled differently, depending on its characteristics. Below are some of the inefficient parts noticed, along with suggested improvements for data preprocessing:

### 1. Lack of Data Exploration and Understanding
- **Issue**: 
  - Before preprocessing, there was insufficient exploratory data analysis (EDA), such as examining the distribution, proportion of missing values, and presence of outliers.
- **Improvement**: 
  - Perform a thorough EDA before data preprocessing. This includes:
    - Analyzing basic statistics
    - Visualizing data distributions
    - Checking for missing values
    - Identifying outliers  
  This helps to create targeted preprocessing strategies and avoid unnecessary operations.

### 2. Simplistic Handling of Missing Values
- **Issue**: 
  - The approach to handling missing values may have been too simple, involving either filling or dropping them without deeper consideration.
- **Improvement**: 
  - For numerical variables, use smarter imputation methods, such as mean, regression, or KNN imputation.
  - For categorical variables, consider predictive imputation models instead of simple mode imputation.
  - Treat missing values as a feature (e.g., add a binary column indicating whether a value is missing) so that the model can learn the impact of missing data.

### 3. Overuse or Neglect of Feature Scaling
- **Issue**: 
  - Feature scaling was either overused or neglected, depending on the model type. Some models (e.g., tree-based models) do not require scaling and may perform worse with it.
- **Improvement**: 
  - Adjust scaling based on the model type:
    - Use standardization or normalization for linear models or K-means clustering.
    - Avoid scaling for tree-based models like decision trees or random forests.
  - Use different scaling methods (e.g., Min-Max Scaling, Robust Scaling) depending on the data distribution to better handle outliers and skewed features.

### 4. Simplistic Text Processing
- **Issue**: 
  - Text processing may have been too basic, using only bag-of-words or TF-IDF models, which do not capture context.
- **Improvement**: 
  - Use more sophisticated text representation techniques like word embeddings (Word2Vec, GloVe) or contextual embeddings (e.g., BERT) to enhance the model's understanding of text data.
  - Implement more comprehensive text cleaning, such as removing punctuation, low/high-frequency words, and applying stemming or lemmatization.

### 5. Insufficient Handling of Imbalanced Classes
- **Issue**: 
  - Imbalanced datasets might have been handled using simple oversampling or undersampling methods, which could lead to overfitting or information loss.
- **Improvement**: 
  - Use more advanced techniques like SMOTE, ADASYN, or class weight adjustments to handle imbalanced data.
  - Consider models that are more sensitive to imbalanced data, such as gradient boosting models (e.g., XGBoost, LightGBM), which can handle imbalanced class distributions better.

### 6. Lack of Feature Selection
- **Issue**: 
  - All features were likely treated equally during preprocessing, without considering whether certain features were irrelevant or harmful to the model.
- **Improvement**: 
  - Perform feature selection during preprocessing using:
    - Correlation analysis
    - Chi-squared tests
    - Model-based feature importance  
  This helps select the most informative features, reduce dimensionality, and improve model performance.
